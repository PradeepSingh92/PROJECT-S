```python
import pandas as pd
import re

def profile_column_cleanliness(df):
    total_rows = len(df)
    report = []

    # Define junk pattern
    junk_pattern = r'^\s*$|^[-.,]+$|^(N/A|null|None)$'

    for col in df.columns:
        # Convert to string and normalize
        cleaned = df[col].astype(str).replace(junk_pattern, pd.NA, regex=True)
        junk_count = cleaned.isna().sum()
        junk_percent = round((junk_count / total_rows) * 100, 2)

        report.append({
            'Column Name': col,
            'Total Rows': total_rows,
            'Junk Count': junk_count,
            'Junk %': junk_percent,
            'Fully Junk': junk_count == total_rows,
            'Mostly Junk': junk_percent > 80 and junk_count < total_rows
        })

    return pd.DataFrame(report)

# Example usage
cleanliness_report = profile_column_cleanliness(data)
print(cleanliness_report)
```
