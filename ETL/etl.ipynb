```python
import pandas as pd

def extract_data(file_path):
    try:
        df = pd.read_csv(file_path)
        print("‚úÖ Data extracted successfully.")
        return df
    except Exception as e:
        print(f"‚ùå Extraction failed: {e}")
        return None

# Test extraction
data = extract_data("/Users/pradeepsingh/Downloads/salesdata.csv")
if data is not None:
    print(data.head())
```
